```python
x = df.drop('Heart Disease Status', axis =1)
y = df['Heart Disease Status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

```python
# testing

conf_mat = confusion_matrix(y_test, y_pred)
print("Condusion Matrx", confusion_matrix)

acc_score = accuracy_score(y_test, y_pred)
print("Accuracy Score", acc_score)

class_report = classification_report(y_test, y_pred)
print("Classification Report", class_report)
```

```python
# training

y_train_pred = rf_model.predict(x_train)

conf_mat = confusion_matrix(y_train, y_train_pred)
print("Condusion Matrx", confusion_matrix)

acc_score = accuracy_score(y_train, y_train_pred)
print("Accuracy Score", acc_score)

class_report = classification_report(y_train, y_train_pred)
print("Classification Report", class_report)
```

# Hyper Parameter Tuning

```python
rf_model = RandomForestClassifier()

parameters = {'criterion': ["gini"],
              'max_depth': np.arange(2, 6),
              'min_samples_split': np.arange(3,8),
              'min_samples_leaf': np.arange(1,5)}
```

```python
gscv = GridSearchCV(rf_model, parameters, cv=5)
gscv.fit(x_train, y_train)
```

```python
rf_hymodel = RandomForestClassifier(max_depth=np.int64(2), min_samples_leaf=np.int64(1),
                       min_samples_split=np.int64(3))
rf_hymodel.fit(x_train, y_train)
```

```python
y_pred = rf_hymodel.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

```python
#testing

con_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix", con_mat)

acc_score = accuracy_score(y_test, y_pred)
print("Acuuracy Score", acc_score)

class_report = classification_report(y_test, y_pred)
print("Classification Report", class_report)
```

```python
#training

y_Pred_train = rf_hymodel.predict(x_train)

con_mat = confusion_matrix(y_train, y_Pred_train)
print("Confusion Matrix", con_mat)

acc_score = accuracy_score(y_train, y_Pred_train)
print("Acuuracy Score", acc_score)

class_report = classification_report(y_train, y_Pred_train)
print("Classification Report", class_report)
```
